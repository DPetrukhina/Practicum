### Предобработка данных

#### Проверка типов данных

```python
data.dtypes # проверим типы данных
```

**Тип данных столбцов `show_start_date` и `ratings` нуждается в изменении. Остальные типы данных соответствуют необходимым.**

```python
# изменим тип данных столбца 'show_start_date'
data['show_start_date'] = pd.to_datetime(data['show_start_date'], format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
# посмотрим, что может быть не так с данными столбца 'ratings'
invalid_raitings = data[data['ratings'].str.contains('[^0-9.]').fillna(False)]
invalid_raitings
```

Часть данных хранится в формате процентов, хотя обычно для рейтингов используют десятибальную систему. Перепишем проценты в этих строках как доли от 10 баллов, а затем приведём всё к числу.

```python
# исправим тип данных в колонке 'ratings'
data.loc[invalid_raitings.index, 'ratings'] = invalid_raitings[
                                'ratings'].str[0] + '.' + invalid_raitings['ratings'].str[1]
data['ratings'] = pd.to_numeric(data['ratings'])
data.dtypes
```

**Рассмотри пропуски в остальных столбцах**

Пропусков в столбце `film_studio` уже намного больше, их искать в Интернете уже не будем. Их, как и пропуски в столбцах `director`, `producer` и `genres` заменим на `неизвестно`(столбцы для замены содержат значения типа `object`).

В нём столько же пропусков, сколько и в столбцах `refundable_support`, `nonrefundable_support`, `financing_source`. Убедимся, что они, действительно совпадают.

```python
data.query(
    'financing_source.isna() & nonrefundable_support.isna() & refundable_support.isna() & budget.isna()')
```

Их слишком много, чтобы удалить, поэтому оставим, где тип данных `вещественный`, а где строковый - заменим так же, как и до этого заменяли пропуски, но позже (для дальнейшего анализа и визуализации могут понадобиться данные в исходном виде).

Далее будут проанализированы фильмы с государственной поддержкой и без неё для выявления аспектов влияния этого параметра на другие параметры, и, наоборот. Будет проведено сравнение фильмов с гос. поодержкой и без неё, динамика изменений сумм гос. поддержки.

Завершающей частью анализа станет оценка рейтинга фильмов, его зависимости с другими параметрами и динамики во времени.

**Таким образом, анализ охватит создание фильмов, интерес к ним зрителя, а также субъективную оценку зрителя. Каждый этап этого анализа позволит создать наиболее полное представление о рынке кинопроката.**


```python
type_box = data[data['box_office'].notna()] \
    .pivot_table(index='year',
                columns='type',
                values='box_office',
                aggfunc='sum')
display(type_box)
# визуализируем данные
type_box.plot(title='Cумма сборов по годам с разбивкой по типам', kind='line', xlim=[
    2010, 2019], style='o-', grid=True, figsize=(15,10))
plt.xlabel('Годы')
plt.ylabel('Сумма сборов')
plt.show()
type_box.reset_index(
).plot(
    x='year', kind='bar', figsize = (10, 5), grid = True, title='Cумма сборов по годам с разбивкой по типам')
plt.xlabel('Годы')
plt.ylabel('Сумма сборов')
plt.show()
```

```python
# визуализируем данные
data['type'].value_counts().plot(y='count', kind='pie', figsize=(8, 8), autopct='%1.1f%%')
plt.title('Распределение фильмов по возрастным категориям')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

**Таким образом, большинство кассовых фильмов относятся к художественным или же анимационным.** Причём, каждый год первые приносят намного больше вторых.
